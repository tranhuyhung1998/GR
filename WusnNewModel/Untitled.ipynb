```python
%matplotlib inline 
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
from matplotlib import cm
import numpy as np
from matplotlib import ticker


dem = DemsInput.from_file('data/dems_data/dem10.asc')
arr = [x[:41] for x in dem.height[:41]]

inp = WusnInput.from_file('data/small_data/no-dem10_r25_1.in')
X = np.clip([x.x for x in inp.sensors + inp.relays], 2, 198)
Y = np.clip([x.y for x in inp.sensors + inp.relays], 2, 198)

# for x,y in zip(X,Y):
#     print(x,y)
figure(num=None, figsize=(8,6), dpi=80, facecolor='w', edgecolor='k')
cp = plt.contourf([x*5 for x in list(range(41))], [x*5 for x in list(range(41))], arr, cmap=cm.YlOrBr, figsize=(500,500))
plt.scatter(x=X, y=Y, color='blue', s=4)
plt.gca().set_aspect('equal', adjustable='box')
cbar = plt.colorbar(cp)
cbar.ax.tick_params(labelsize=14)


# (generate plot here)
# tick_locator = ticker.MaxNLocator(nbins=5)
# cbar.locator = tick_locator
# cbar.update_ticks()
cbar.set_ticks([2560, 2800, 3040, 3280])

plt.xticks()
plt.xticks(fontsize=14)
plt.yticks(np.arange(0, 201, 50))
plt.yticks(fontsize=14)
plt.savefig('no.eps', format='eps', dpi=1000)
```
